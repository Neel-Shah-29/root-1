```python
def generator(array1, array2 ,batch_size):
  inputs = []
  outputs = []
  batchcount = 0
  while True:
       for x in array1:
         inputs.append(x)
       for x1 in array2:
         outputs.append(x1)
         batchcount += 1
         if batchcount > batch_size:
                  X = np.array(inputs, dtype='float32')
                  y = np.array(outputs, dtype='float32')
                  yield (X,y)
                  inputs = []
                  outputs = []
                  batchcount = 0
```

```python
#Total number of events in a Branch of a signal Tree
Nevents=10000
```

```python
from keras.models import Sequential
from keras.layers import Dense,Flatten
model = Sequential()
model.add(Dense(12, input_dim=10000, activation="relu"))
model.add(Dense(12, activation="relu"))
model.add(Dense(1, activation="sigmoid"))
model.add(Flatten())
model.summary()
model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```
